```
#@title Imports
import tensorflow as tf
from google.colab import files
from google.protobuf.json_format import MessageToJson
import json
```

```
#@title Upload tfrecord examples
uploaded = files.upload()
```

```
#@title View TFRecord
raw_dataset = tf.data.TFRecordDataset("b1c7.tfrecord")

for raw_record in raw_dataset.take(1):
    example = tf.train.Example()
    example.ParseFromString(raw_record.numpy())
    print(example)
    #jsonMessage = MessageToJson(example)
    #print(jsonMessage)
```

```
#@title View items in TFRecord
kv = example.features.feature.keys()
print(len(kv))
for item in kv:
  print(item)
```

```
#@title Convert TFRecord to JSON and Display and save.
input_file = "b1c7.tfrecord"
raw_dataset = tf.data.TFRecordDataset(input_file)
output_path = input_file.split('.')[0]+".json"

for raw_record in raw_dataset.take(1):
    example = tf.train.Example()
    example.ParseFromString(raw_record.numpy())
    jsonMessage = MessageToJson(example)
    print(jsonMessage)
    with open(output_path, "w") as opfd:
      opfd.write(jsonMessage)
```

```
#@title Download json file
files.download(output_path)
```
